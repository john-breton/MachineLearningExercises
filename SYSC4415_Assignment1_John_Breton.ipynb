##Question 1 i)
> In order to calculate the gradient of a function, we need to take the partial derivative with respect to each of the inputs of the function. In this case, the function has three inputs: $ x,y,z $. This can be demonstrated a little more concisely by denoting the gradient vector of the above function (i.e., $ \nabla(f) $): $$ \nabla(f) = \begin{bmatrix}\frac{\partial}{\partial\:x}(f)\\ \frac{\partial}{\partial\:y}(f)\\ \frac{\partial}{\partial\:z}(f)\end{bmatrix} = \begin{bmatrix}\frac{\partial}{\partial\:x}(z^3+x^2y-y^2+3yz)\\ \frac{\partial}{\partial\:y}(z^3+x^2y-y^2+3yz)\\ \frac{\partial}{\partial\:z}(z^3+x^2y-y^2+3yz)\end{bmatrix} $$ <br/> We can then evaluate each element of the matrix individually to determine our final gradient vector. Beginning with $\frac{\partial}{\partial\:x}(f)$:
$$\frac{\partial}{\partial\:x}(z^3+x^2y-y^2+3yz) = 0 + 2xy - 0 + 0$$
$$\frac{\partial}{\partial\:x}(z^3+x^2y-y^2+3yz) = 2xy $$
<br/> Next with $\frac{\partial}{\partial\:y}(f)$:
$$\frac{\partial}{\partial\:y}(z^3+x^2y-y^2+3yz) = 0 + x^2 - 2y + 3z $$
$$\frac{\partial}{\partial\:y}(z^3+x^2y-y^2+3yz) = x^2 -2y + 3z $$
<br/> Lastly we compute $\frac{\partial}{\partial\:z}(f)$:
$$\frac{\partial}{\partial\:z}(z^3+x^2y-y^2+3yz) = 3z^2 + 0 - 0 + 3y $$
$$\frac{\partial}{\partial\:z}(z^3+x^2y-y^2+3yz) = 3z^2 + 3y $$
<br/> Finally, we can rewrite our gradient vector with this calculated partial derivatives: $$\therefore\:\nabla(f) = \begin{bmatrix}2xy \\ x^2 -2y + 3z \\ 3z^2 + 3y\end{bmatrix}$$

##Question 1 ii)
> We can calculate the gradient at (-2, 3, 1) using our results for i):$$\nabla(f(-2, 3, 1)) = \begin{bmatrix}2*-2*3 \\ (-2)^2 -2(3) + 3(1) \\ 3(1)^2 + 3(3)\end{bmatrix}$$</br>
$$\therefore\:\nabla(f(-2, 3, 1)) = \begin{bmatrix}-12 \\ 1 \\ 12\end{bmatrix}$$

##Question 1 iii)
> The result of the gradient at any point will be a trajectory towards the greatest increase of the function. In other words, the resulting vector is a direction to a point in the function where the gradient is 0 (or a local maximum for the function). We can't be certain if this is the absolute maximum of the function because there could be multiple points at which the gradient is 0 (multiple local maximums). However, in this case all we know is the result of the computation for ii) is a trajectory pointing us to the closest maximum of the function relative to the point we passed in (-2, 3, 1).

##Question 2 a.
>We are asked to calculate three values. The expected value from this sample, along with the sample variance and standard deviation using an unbiased estimator. We can start with the expected value of the sample. It might be easier to display the results of the survey as a table showing the number of times a each respond was chosen:

Hours of Sleep | Responses | Pr(Hours of Sleep)
------ | ------ | ------
1   | 3 | 0.3
2   | 3 | 0.3
3   | 2 | 0.2
4   | 1 | 0.1
5   | 1 | 0.1

>Since there are 10 responses, assigning a probability to each response type is made easy. The third column in the table above shows the likelihood of a response being chosen given the sample presented in a question. Knowning this information, we can now calculate the expected value (EV) from this sample:
$$ EV(x) = \sum Pr(x)\times x $$
$$ EV(x) = 1(0.3) + 2(0.3) + 3(0.2) + 4(0.1) + 5(0.1) $$
$$ EV(x) = 2.4 $$
Next, we can calculate the sample variance using an unbiased estimator (so we need to use $(n-1)$ rather than just $n$):
$$S^2=\frac{\sum(x_i - \bar{x})^2} {n-1}$$
$$\bar{x} = \frac{(1 + 1 + 1 + 2 + 2 + 2 + 3 + 3 + 4 + 5)}{10}$$
$$\bar{x} = 2.4$$
$$S^2=\frac{(1-2.4)^2 + (1-2.4)^2 + (1-2.4)^2 + (2-2.4)^2 + (2-2.4)^2 + (2-2.4)^2 + (3-2.4)^2 + (3-2.4)^2 + (4-2.4)^2 + (5-2.4)^2}{10 - 1}$$
$$S^2=1.8\overline{22}$$
Lastly, we calculate the standard deviation from our unbiased sample variance:
$$S=\sqrt{1.8\overline{22}}$$
$$S\approx1.3499$$

$\therefore$ Our expected value from this sample is 2.4, our sample variance using an unbiased estimator is $1.8\overline{22}$, and our standard deviation is approximately 1.3499.

##Quetion 2 c.
> We can use the same equations from Q2a to solve for the expected value and the variance of Pr(x), which in this case are the values of a pmf:
$$ EV(x) = \sum Pr(x)\times x $$
$$ EV(x) = 1(0.25) + 2(0.3) + 3(0.2) + 4(0.1) + 5(0.15) $$

Next is the variance of Pr(x), note the equation is different this time since we are working with a pmf:

##Question 2 d.
>In order to make use of Bayes' Theorem, we first need to calculat $Pr(-)$. Fortunately, we have the tools necessary to determine $Pr(-)$, as we can calculate $Pr(+)$ and then make use of the relationship that $Pr(-) = 1 - Pr(+)$.
$Pr(+)$ is defined as:
$$\sum Pr(+|x)Pr(x)$$
Our provided data has all of these necessary values (save for Pr(3) which we previously determined above), thus we end up with the following:
$$Pr(+) = (0.25)(0.75) + (0.3)(0.55) + (0.2)(0.35) + (0.1)(0.2) + (0.15)(0.1)$$
$$Pr(+) = 0.4575$$
Subsequently:
$$Pr(-) = 1 - Pr(+)$$
$$Pr(-) = 0.5425$$
The only thing left to determine before we can make use of Bayes' Theorem is $Pr(-|x = 2)$, as we are trying to determine the probability that a student slept 2 hours the night before. Fortunately this is another simple subtraction:
$$Pr(-|x = 2) = 1 - Pr(+|x = 2)$$
$$Pr(-|x = 2) = 1 - 0.55$$
$$Pr(-|x = 2) = 0.45$$
Finally, we can plug and chug into Bayes' Theorem and determine $Pr(2|-)$:
$$Pr(x|-)=\frac{Pr(-|x)Pr(x)}{Pr(-)}$$
$$Pr(2|-)=\frac{Pr(-|2)Pr(2)}{Pr(-)}$$
$$Pr(2|-)=\frac{(0.45)(0.3)}{0.5425}$$
$$Pr(2|-)\approx0.2488$$

$\therefore$ The probability that a student slept 2 hours the night before given that they did not need excessive amounts of coffe is roughly 24.88%.

## Question 3 a) Create the dataset 

The first step is loading the Kaggle Diabetes data. We will then split off the test data to be used for all training sets. Then create each training set, using **stratified sampling**

```python
# Load the Kaggle Diabetes dataset 
# (This is assuming it had been imported into the notebook)
kaggle_data = pd.read_csv("/content/diabetes.csv")
df = pd.DataFrame(kaggle_data)

# Fortunately all of the data is numerical so we don't need to do transforms.
# Prepare x and y
# Extract out data (minus our dependent var):
x = df[['Pregnancies', 'Glucose','BloodPressure', 'SkinThickness', 'Insulin',
        'BMI', 'DiabetesPedigreeFunction', 'Age']]
x = sm.add_constant(x)
# Get our dependent var
y = df['Outcome']

# Partition training and testing data sets (75/25 split) 
# Making sure to stratify the data, and using random_state=0 so it is
# consistent when I discuss the results of the model accuracy below.
train_x, test_x, train_y, test_y = train_test_split(x, y, test_size=.25, 
                                                    random_state=0, stratify=y)
```

## Question 4 b) c) d) e)
## Step 2: Linear model $y=mx+b$

```python
# Transform y so we can acutally use it in calculations
y_arr = np.array(y_plot)

# Augment the x vector
x_lin_aug = np.concatenate([np.ones((len(x_plot), 1)), x_plot], axis=1)
x_lin_aug_arr = np.array(x_lin_aug)

# Compute beta
temp_1 = np.linalg.inv(np.matmul(x_lin_aug_arr.T, x_lin_aug_arr))
temp_2 = np.matmul(x_lin_aug_arr.T, y_arr)
beta = np.matmul(temp_1, temp_2)
print("Calculated beta values = [{}, {}]".format(beta[0][0], beta[1][0]))
print("This means that in our linear model m = {} and b = {}".format(beta[0][0], 
                                                                     beta[1][0]))

# Calculate the line of best fit
x_line = np.linspace(-14, 19, 150)
y_line = [0]*150
for i in range(len(x_line)):
    y_line[i] = (beta[1][0] * x_line[i]) + beta[0][0]

# Plot the data with the line of best fit.
plt.figure(figsize= (7, 7))
plt.scatter(x_plot, y_plot, label="Dataset")
plt.plot(x_line, y_line, color="purple", label="Linear LOBF")
plt.xlabel("First Column [0] - X")
plt.ylabel("Second Column [1] - Y")
plt.title("Assig1Q4 Dataset - Linear")
plt.legend()
plt.show()

# Compute the MSE
MSE = 0
for i in range(len(x_plot)):
    MSE += (y_arr[i] - ((beta[1][0] * x_plot[0][i]) + beta[0][0]))**2
print("MSE =", (MSE/len(x_plot))[0])
```

## Question 4 f)
## Step 3: Cubic model $y=ax^3+bx^2+cx+d$

```python
# Augment the x vector
# We are calculating 4 unknowns in a cubic equation, so we need 4 columns,
# with 3 being our data exponentiated when necessary. 
x_cubic_aug = np.concatenate([np.ones((len(x_plot), 1)), 
                              x_plot, x_plot ** 2, x_plot ** 3], axis=1)
x_cubic_aug_arr = np.array(x_cubic_aug)

# Compute beta
temp_1_cubic = np.linalg.inv(np.matmul(x_cubic_aug_arr.T, x_cubic_aug_arr))
temp_2_cubic = np.matmul(x_cubic_aug_arr.T, y_arr)
beta_cubic = np.matmul(temp_1_cubic, temp_2_cubic)
print("Calculated beta values = [{}, {}, {}, {}]".format(beta_cubic[0][0], 
                                                         beta_cubic[1][0],
                                                         beta_cubic[2][0],
                                                         beta_cubic[3][0]))
print("This means that in our cubic model a = {}, b = {}, c = {}, and d = {}"
.format(beta_cubic[3][0], beta_cubic[2][0], beta_cubic[1][0], beta_cubic[0][0]))

# Calculate the line of best fit
x_line_cubic = np.linspace(-14, 19, 150)
y_line_cubic = [0]*150
for i in range(len(x_line_cubic)):
    y_line_cubic[i] = (beta_cubic[3][0] * x_line_cubic[i] ** 3) + \
                      (beta_cubic[2][0] * x_line_cubic[i] ** 2) + \
                      (beta_cubic[1][0] * x_line_cubic[i] ** 1) + \
                       beta_cubic[0][0]

# Plot the data with the line of best fit.
plt.figure(figsize= (7, 7))
plt.scatter(x_plot, y_plot, label="Dataset")
plt.plot(x_line_cubic, y_line_cubic, color="purple", label="Cubic LOBF")
plt.xlabel("First Column [0] - X")
plt.ylabel("Second Column [1] - Y")
plt.title("Assig1Q4 Dataset - Cubic")
plt.legend()
plt.show()

# Compute the MSE
MSE_quad = 0
for i in range(len(x_plot)):
    MSE_quad += (y_arr[i] - ((beta_cubic[3][0] * x_line_cubic[i] ** 3) + \
                             (beta_cubic[2][0] * x_line_cubic[i] ** 2) + \
                             (beta_cubic[1][0] * x_line_cubic[i] ** 1) + \
                             beta_cubic[0][0]))**2
print("MSE =", (MSE_quad/len(x_plot))[0])
```
